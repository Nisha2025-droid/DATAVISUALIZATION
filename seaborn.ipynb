```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the example dataset
tips = sns.load_dataset("tips")
```

```python
tips.head(4)
```

```python
tips['size'].unique()
```

```python
sns.distplot(tips['total_bill'])
```

```python
plt.subplot(1, 2, 1)
sns.distplot(tips['total_bill'], kde=False, bins=30)
plt.subplot(1, 2, 2)
sns.distplot(tips['total_bill'], kde=True, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,kind = 'scatter')
```

```python
sns.pairplot(tips, hue='sex',palette='rainbow')
```

```python
df = sns.load_dataset('tips')
```

```python
sns.countplot(x= df['sex'],hue = df['smoker'])
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,palette='rainbow')   
```
