```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,5,10)
y= x**2
x
y
```

```python
plt.plot(x,y)
plt.title("relationship between x and y")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
plt.subplots(2,2)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)   
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])    
ax.plot(x,y)
ax.set_title("relationship between x and y")
```
